```python
import time 
import psutil
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation 
from matplotlib import style

%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'

#Clear text file
'''file = open("text.txt","r+")
file.truncate(0)
file.close()'''

#Real time Plot
plt.style.use('ggplot')
fig = plt.figure(figsize=(6,8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
fig.show()
```

```python
while True:   
    
    reward = []
    data = open("Plot_model_1.txt", "r").read()
    line = data.split('\n')[:-1]
    x = np.arange(1,len(line)+1)

    for i in range(0, len(line)):
        if len(line) >= 1:
            reward.append(float(line[i]))

    average_reward = np.cumsum(reward)/x
    
    ax1.plot(x, reward, color='b', linewidth=1)
    ax1.title.set_text('Reward')
    
    ax2.plot(x, average_reward, color='b', linewidth=1)
    ax2.title.set_text('Average Reward')
    
    fig.canvas.draw()
    
    time.sleep(5)   
```
